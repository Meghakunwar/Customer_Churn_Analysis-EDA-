# replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

# converted 0 and 1 value of senior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x="SeniorCitizen", data=df, hue="Churn")
plt.title("Churn by SeniorCitizen")
plt.show()
```

```python

# CLEAN COLUMN NAMES
# =========================
df.columns = df.columns.str.strip()

# =========================
# COUNT DATA
# =========================
count_df = (
    df.groupby(['SeniorCitizen', 'Churn'])
      .size()
      .reset_index(name='count')
)

# =========================
# CALCULATE PERCENTAGE (FIXED)
# =========================
count_df['percent'] = (
    count_df['count']
    / count_df.groupby('SeniorCitizen')['count'].transform('sum')
    * 100
)

# =========================
# CREATE PIVOT TABLE
# =========================
pivot_df = count_df.pivot(
    index='SeniorCitizen',
    columns='Churn',
    values='percent'
)

# =========================
# PLOT STACKED BAR CHART
# =========================
ax = pivot_df.plot(
    kind='bar',
    stacked=True,
    figsize=(3,4)
)

# =========================
# ADD PERCENT LABELS
# =========================
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

# =========================
# FORMATTING
# =========================
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)')
plt.title('Churn by SeniorCitizen (%)')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
# ===============================
# 1. Import Libraries
# ===============================

import math

# ===============================

# 3. Fix Data Types
# ===============================
# Convert TotalCharges to numeric (common issue in Telco dataset)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# ===============================
# 4. Select Columns for Countplots
# ===============================
# Exclude continuous numeric columns
exclude_cols = ['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']

categorical_cols = [
    col for col in df.columns
    if col not in exclude_cols
]

# ===============================
# 5. Set Plot Grid Size
# ===============================
n_cols = 3
n_rows = math.ceil(len(categorical_cols) / n_cols)

plt.figure(figsize=(20, n_rows * 4))

# ===============================
# 6. Create Countplot Subplots
# ===============================
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    sns.countplot(
        x=col,
        data=df,hue=df["Churn"],
        order=df[col].value_counts().index
    )
    plt.title(f'Count Plot of {col}')
    plt.xticks(rotation=45)

# ===============================
# 7. Layout Adjustment
# ===============================
plt.tight_layout()
plt.show()
```

# The countplots show that customers without add-on services (OnlineSecurity, TechSupport, DeviceProtection) have a much higher churn rate.
# Month-to-month contracts, paperless billing, and electronic check payments are strongly associated with higher churn.
# Customers with longer contracts (1–2 years) and support services enabled tend to stay longer.
# Overall, churn is more common among newer customers with fewer bundled services.
